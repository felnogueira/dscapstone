```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import MarkerCluster
# Import folium MousePosition plugin
from folium.plugins import MousePosition
# Import folium DivIcon plugin
from folium.features import DivIcon
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv')
```

```python
spacex_df = df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
spacex_map = folium.Map(location=[39.8283,-98.5795],zoom_start=5,tiles="Cartodb positron")
```

```python
spacex_map
```

```python
for launch_site, lat, long in zip(launch_sites_df["Launch Site"],launch_sites_df.Lat,launch_sites_df.Long):
    circle = folium.Circle([lat,long],radius=1000,color = "red", fill=True).add_child(folium.Popup(launch_site))

    marker=folium.map.Marker(
        [lat,long],

        icon = DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html = f'<div style="font-size: 12; color:#d35400;"><b>{launch_site}</b></div>'
        )
    )

    spacex_map.add_child(circle)
    spacex_map.add_child(marker)
```

```python
spacex_df["marker_color"] = spacex_df["class"].map({0:"red",1:"green"})
```

```python
spacex_df["outcome"] = spacex_df["class"].map({0:"Fail",1:"Success"})
```

```python
marker_cluster = MarkerCluster()
for lat, long, color, outcome in zip(spacex_df["Lat"],spacex_df["Long"],spacex_df["marker_color"],spacex_df["outcome"]):
    marker = folium.Marker(
        [lat,long],
        icon=DivIcon(
            html=f'<div style="font-size: 12; color:{color};"><b>{outcome}</b></div>'
        )
    )
    marker_cluster.add_child(marker)
```

```python
spacex_map.add_child(marker_cluster)
```
