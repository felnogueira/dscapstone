```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
```

```python
df = pd.read_csv(url)
```

```python
df.info()
```

```python
plt.figure(figsize = (20,10))
plt.scatter(df["FlightNumber"],df["PayloadMass"],c=df["Class"], alpha=0.5, marker = "o")
plt.xlabel("FlightNumber")
plt.ylabel("PayloadMass")
plt.show()
```

```python
df
```

```python
plt.figure(figsize=(20, 10))
# Assign the scatter to a variable (scatter)
scatter = plt.scatter(df["FlightNumber"], df["PayloadMass"], c=df["Class"], alpha=0.5, marker="o")

plt.xlabel("FlightNumber")
plt.ylabel("PayloadMass")

# Add the legend using the scatter object's elements
plt.legend(*scatter.legend_elements(), title="Class")

plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(df,x="FlightNumber",y="PayloadMass",hue="Class")
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
orbit_success = df.groupby("Orbit")["Class"].sum().rename("Class_sum").to_frame().reset_index()
```

```python
orbit_success.sort_values(by="Class_sum",ascending=False,inplace=True)
```

```python
orbit_success
```

```python
plt.figure(figsize=(20,10))
sns.barplot(orbit_success,x="Orbit",y="Class_sum")
```

## mean


```python
orbit_rate = df.groupby("Orbit")["Class"].mean().rename("SuccessRate").reset_index()
```

```python
orbit_rate.sort_values(by="SuccessRate",ascending=False,inplace=True)
```

```python
orbit_rate["SuccessRate"] = orbit_rate["SuccessRate"] * 100
```

```python
plt.figure(figsize=(20,10))
sns.barplot(orbit_rate,x="Orbit",y="SuccessRate")
plt.ylabel("Success Rate (%)")
```
