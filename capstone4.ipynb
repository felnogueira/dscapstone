```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
```

```python
df.to_sql("SPACEXTBL",con, if_exists="replace",index=False,method="multi")
```

```python
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
```

```python
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null
```

```python
%%sql
SELECT * FROM SPACEXTABLE LIMIT 5
```

```python
%%sql
SELECT DISTINCT Launch_Site from SPACEXTABLE
```

```python
%%sql
SELECT * FROM SPACEXTABLE
WHERE Launch_Site LIKE "CCA%" LIMIT 5
```

```python
%%sql
SELECT SUM(PAYLOAD_MASS__KG_) FROM SPACEXTABLE
WHERE CUSTOMER LIKE "NASA (CRS)"
```

```python
%%sql
SELECT AVG(PAYLOAD_MASS__KG_) FROM SPACEXTABLE
WHERE Booster_Version LIKE "F9 v1%"
```

```python
%%sql
SELECT DISTINCT Landing_Outcome FROM SPACEXTABLE
```

```python
%%sql
SELECT MIN(DATE) FROM SPACEXTABLE
WHERE Landing_Outcome = "Success (ground pad)"
```

```python
%%sql
SELECT DISTINCT Booster_Version FROM SPACEXTABLE
WHERE Landing_Outcome = "Success (drone ship)" AND
PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000
```

```python
%%sql
SELECT * FROM SPACEXTABLE LIMIT 5
```

```python
%%sql
SELECT Mission_Outcome, COUNT(*) FROM SPACEXTABLE
GROUP BY Mission_Outcome
```

```python
%%sql
SELECT DISTINCT Booster_Version FROM SPACEXTABLE
WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTABLE)
```

```python
%%sql
SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTABLE
```

```python
%%sql
SELECT SUBSTR(Date,6,2) AS "Month", Landing_Outcome, Booster_Version, Launch_Site FROM SPACEXTABLE
WHERE SUBSTR(Date,0,5) = "2015" AND Landing_Outcome = "Failure (drone ship)"
```

```python
%%sql
SELECT DISTINCT Landing_Outcome FROM SPACEXTABLE
```

```python
%%sql
SELECT Landing_Outcome, COUNT(*) FROM SPACEXTABLE
WHERE Date > "2010-06-04" AND Date < "2017-03-20"
GROUP BY Landing_Outcome
ORDER BY COUNT(*) DESC
```

```python
%%sql
SELECT Landing_Outcome, COUNT(*) FROM SPACEXTABLE
GROUP BY Landing_Outcome
ORDER BY COUNT(*) DESC
```
